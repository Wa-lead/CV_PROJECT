```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


ss = pd.read_csv('data/pcba_all.csv')
ss.head()
```

```python
mols = list(ss.columns[6:])
columns = ['rdkit SMILES'] + mols

ss = ss.dropna(subset=['rdkit SMILES'])
df = ss[columns]
```

```python
df_selected = df[['rdkit SMILES','pcba-aid686978' ]]
df_selected = df_selected.rename(columns={'rdkit SMILES': 'smiles', 'pcba-aid686978': 'label'})
```

```python
df_selected_1 = df_selected[df_selected['label'] == 1]
df_selected_1 = df_selected_1.sample(frac=0.7)

df_selected_0 = df_selected[df_selected['label'] == 0]
df_selected_0 = df_selected_0.sample(len(df_selected_1))

df_selected = pd.concat([df_selected_1, df_selected_0])
df_selected
```

```python
train_df, test_df = train_test_split(df_selected, test_size=0.3, random_state=42, stratify=df_selected['label'])
val_df, test_df = train_test_split(test_df, test_size=0.5, random_state=42, stratify=test_df['label'])

train_df.to_csv('train_aid686978_0.7.csv', index=False)
val_df.to_csv('val_aid686978_0.7.csv', index=False)
test_df.to_csv('test_aid686978_0.7.csv', index=False)
```
